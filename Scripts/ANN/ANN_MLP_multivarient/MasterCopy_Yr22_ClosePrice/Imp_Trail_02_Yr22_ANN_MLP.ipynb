```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/IIIT_Guwahati/ANN_MLP/Yr_22/PrepStep_TCS.2002 sept_NS.csv'  # Update this path if necessary
data = pd.read_csv(file_path, parse_dates=['Date'], dayfirst=True)

# Display the original dataset
print("Original dataset:")
print(data)

# Remove the last 15 rows
data = data[:-25]

# Display the updated dataset
print("\nUpdated dataset (last 10 rows removed):")
print(data)

# Save the updated dataset to a new CSV file if needed
# data.to_csv('/mnt/data/TCS_News_Dataset_Updated.csv', index=False)

```

```python
# Convert the 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Sort data by date
data = data.sort_values('Date')

# Fill missing values
data = data.fillna(method='ffill')

# Extract the 'Close' prices
closing_prices = data['Close'].values
```

## step2: Normalization

```python
# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
closing_prices = closing_prices.reshape(-1, 1)
closing_prices_scaled = scaler.fit_transform(closing_prices)
```

## step3: Model_creating

```python
# Create sequences for the MLP model
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        sequences.append(data[i:i + sequence_length + 1])
    return np.array(sequences)

sequence_length = 10
sequences = create_sequences(closing_prices_scaled, sequence_length)

# Split into features (X) and labels (y)
X = sequences[:, :-1]
y = sequences[:, -1]

# Split into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

### s(i) Train_Model

```python
# Build the MLP model
model = Sequential()
# Input layer and first hidden layer
model.add(Dense(64, input_dim=sequence_length, activation='relu'))
# Second hidden layer
model.add(Dense(32, activation='relu'))
# Output layer
model.add(Dense(1))



# Compile the model, including the 'accuracy' metric
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Train the model (remove 'metrics' from here)
model.fit(X_train, y_train, epochs=200, batch_size=16, validation_split=0.2)
```

## step4: Prediction

```python
# Forecast the next 5 days
last_sequence = closing_prices_scaled[-sequence_length:]
predictions = []
for _ in range(5):
    pred = model.predict(last_sequence.reshape(1, -1))
    predictions.append(pred[0, 0])
    last_sequence = np.append(last_sequence[1:], pred)

# Inverse transform the predictions
predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))
```

```python
# Plot the results
plt.figure(figsize=(40, 15))
plt.plot(data['Date'], closing_prices, label='Actual Closing Prices')
plt.plot(pd.date_range(data['Date'].iloc[-1], periods=6, freq='D')[1:], predictions, label='Predicted Closing Prices', color='red')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('TCS Closing Price Prediction')
plt.legend()
plt.show()
```
